```python
#Q1. Display top and bottom 5 records of the dataset:

import pandas as pd

data = {'X': [0, 1, 2, 3, 4], 'Y': [2, 3, 5, 4, 6]}
df = pd.DataFrame(data)

# Display the top 5 records
print("Top 5 records:")
print(df.head())

# Display the bottom 5 records
print("\nBottom 5 records:")
print(df.tail())
```

```python
#Q2. Display statistical analysis

print("Statistical Analysis:")
print(df.describe())
```

```python
#Q3. Display column names and shape of the dataset

# Display column names
print("Column Names:", df.columns)

# Display shape
print("Shape:", df.shape)
```

```python
#Q4. Display the pair plot of the dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Display pair plot
sns.pairplot(df)
plt.show()
```

```python
#Q5. Find the correlation matrix for the dataset

import pandas as pd

data = {'X': [0, 1, 2, 3, 4], 'Y': [2, 3, 5, 4, 6]}
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
#Q6. Plot the correlation plot on the dataset

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the correlation plot
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Plot")
plt.show()
```

```python
#Q7. Apply linear regression model techniques to predict the dataset

import numpy as np

# Given data
x = np.array([0, 1, 2, 3, 4])
y = np.array([2, 3, 5, 4, 6])

# Calculate the required sums
N = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x_squared = np.sum(x**2)

# Calculate the slope (m) and intercept (c)
m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x**2)
c = (sum_y - m * sum_x) / N

# Print the slope and intercept
print("Slope (m):", m)
print("Intercept (c):", c)
```
