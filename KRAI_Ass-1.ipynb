```python
import pandas as pd

data = {'A': [45, 37, 42, 35, 39], 'B': [38, 31, 26, 28, 33], 'C': [10, 15, 17, 21, 12]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
#Plot the correlation plot on the iris dataset.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

correlation_matrix_iris = iris_df.corr()
sns.heatmap(correlation_matrix_iris, annot=True)
plt.title("Correlation Plot for Iris Dataset")
plt.show()
```

```python
#Analysis of covariance (ANOVA) for iris dataset.
from sklearn.feature_selection import f_classif
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
f_values, p_values = f_classif(X[:, 0].reshape(-1, 1), y)
print("F-values:", f_values[0])
print("P-values:", p_values[0])
```

```python
#Display the pair plot of the iris dataset.
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Create and display the pair plot
sns.set(style="ticks")
sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"])
plt.show()

```
